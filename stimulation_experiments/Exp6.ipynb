```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import resample, sawtooth

# Constants for digital signal
ON_SIGNAL = 8  # microvolts
OFF_SIGNAL = 0  # microvolts
BITS_PER_FEATURE = 16
FORMULA_ANSWER_DURATION = 0.25  # 250 ms for formula and answer
RESPONSE_DURATION = 0.25  # 250 ms for sine/chaos or no stimulus
QUESTION_DURATION = 0.25  # 250 ms for question
NO_STIM_DURATION = 0.25  # 250 ms for no stim
FREQUENCY = 50  # 50 Hz for the sine wave

# New sampling frequency
NEW_FS = 400  # e.g., 400 Hz for a higher sampling rate
NEW_BIT_DURATION = 1 / NEW_FS

def number_to_binary_waveform(number, duration, fs, high_amp=8, low_amp=0):
    """Converts a number to a 16-bit binary waveform."""
    binary_str = format(number, '016b')
    t = np.linspace(0, duration, int(fs * duration), endpoint=False)
    bit_duration = len(t) // 16
    waveform = np.concatenate([(high_amp if bit == '1' else low_amp) * np.ones(bit_duration) for bit in binary_str])
    return waveform

# Plus waveform (Transition from 3.5-4.5 to 0-8)
t_binary = np.linspace(0, FORMULA_ANSWER_DURATION, int(NEW_FS * FORMULA_ANSWER_DURATION), endpoint=False)
y_initial_plus = 0.5 * np.sin(2 * np.pi * FREQUENCY * t_binary) + 4
y_final_plus = 4 * np.sin(2 * np.pi * FREQUENCY * t_binary) + 4
transition_plus = np.linspace(0, 1, len(t_binary))
y_combined_plus = (1 - transition_plus) * y_initial_plus + transition_plus * y_final_plus

# Equals waveform (Sawtooth wave oscillating between 3 and 5)
y_equals = sawtooth(2 * np.pi * FREQUENCY * t_binary)
y_equals = (y_equals + 1) * 1 + 3

# Function to generate sine or chaotic waveform
def generate_waveform(outcome, duration=RESPONSE_DURATION):
    num_samples = int(duration / NEW_BIT_DURATION)
    if outcome == "reward":
        t = np.linspace(0, duration, num_samples, endpoint=False)
        waveform = 4 * (np.sin(2 * np.pi * FREQUENCY * t) + 1)  # Clean sine wave between 0 and 8
    elif outcome == "distress":
        waveform = np.random.uniform(0, 8, num_samples)  # Chaotic wave between 0 and 8
    else:
        waveform = [OFF_SIGNAL] * num_samples  # No signal
    return waveform.tolist()

# Resample waveform
def resample_waveform(waveform, original_fs, target_fs):
    num_samples = int(len(waveform) * target_fs / original_fs)
    return resample(waveform, num_samples)

# Create alternating dataset with arithmetic problems
def create_alternating_dataset(arithmetic_problems):
    dataset = []
    for problem, correct_result in arithmetic_problems:
        operand1, operand2 = map(int, problem.split('+'))
        binary_operand1 = number_to_binary_waveform(operand1, FORMULA_ANSWER_DURATION, NEW_FS)
        binary_operand2 = number_to_binary_waveform(operand2, FORMULA_ANSWER_DURATION, NEW_FS)
        
        # Correct case
        binary_correct_result = number_to_binary_waveform(correct_result, FORMULA_ANSWER_DURATION, NEW_FS)
        combined_correct_binary = np.concatenate((binary_operand1, y_combined_plus, binary_operand2, y_equals, binary_correct_result))
        waveform_correct = generate_waveform("reward", RESPONSE_DURATION)
        stim_signals_correct = np.concatenate((combined_correct_binary, waveform_correct))
        stim_signals_correct = resample_waveform(stim_signals_correct, NEW_FS, NEW_FS)
        dataset.append({
            "problem": problem,
            "result": correct_result,
            "correct": True,
            "binary_operand1": binary_operand1,
            "binary_operand2": binary_operand2,
            "binary_result": binary_correct_result,
            "digital_signals": combined_correct_binary,
            "outcome": "reward",
            "waveform": waveform_correct,
            "stim_signals": stim_signals_correct
        })

        # Incorrect case
        incorrect_result = correct_result - 1 if correct_result > 0 else correct_result + 1
        binary_incorrect_result = number_to_binary_waveform(incorrect_result, FORMULA_ANSWER_DURATION, NEW_FS)
        combined_incorrect_binary = np.concatenate((binary_operand1, y_combined_plus, binary_operand2, y_equals, binary_incorrect_result))
        waveform_incorrect = generate_waveform("distress", RESPONSE_DURATION)
        stim_signals_incorrect = np.concatenate((combined_incorrect_binary, waveform_incorrect))
        stim_signals_incorrect = resample_waveform(stim_signals_incorrect, NEW_FS, NEW_FS)
        dataset.append({
            "problem": problem,
            "result": incorrect_result,
            "correct": False,
            "binary_operand1": binary_operand1,
            "binary_operand2": binary_operand2,
            "binary_result": binary_incorrect_result,
            "digital_signals": combined_incorrect_binary,
            "outcome": "distress",
            "waveform": waveform_incorrect,
            "stim_signals": stim_signals_incorrect
        })

    return dataset

# Create testing dataset based on arithmetic questions
def create_testing_dataset(arithmetic_questions):
    dataset = []
    for question in arithmetic_questions:
        operand1, operand2 = map(int, question.split('+'))
        binary_operand1 = number_to_binary_waveform(operand1, QUESTION_DURATION, NEW_FS)
        binary_operand2 = number_to_binary_waveform(operand2, QUESTION_DURATION, NEW_FS)
        no_stim = [OFF_SIGNAL] * int(NO_STIM_DURATION / NEW_BIT_DURATION)
        question_stim = np.concatenate((binary_operand1, y_combined_plus, binary_operand2, no_stim))
        question_stim = resample_waveform(question_stim, NEW_FS, NEW_FS)
        dataset.append({
            "question": question,
            "binary_operand1": binary_operand1,
            "binary_operand2": binary_operand2,
            "digital_signals": question_stim,
            "question_stim": question_stim
        })
    return dataset

# Generate arithmetic problems and questions from "1 + 1" to "10 + 10"
arithmetic_problems = [(f"{i} + {j}", i + j) for i in range(1, 11) for j in range(1, 11)]
arithmetic_questions = [f"{i} + {j}" for i in range(1, 11) for j in range(1, 11)]

# Create the alternating dataset and testing datasets
alternating_dataset = create_alternating_dataset(arithmetic_problems)
testing_dataset = create_testing_dataset(arithmetic_questions)

# Visualize the dataset as a waveform
def visualize_waveforms(training_dataset, testing_dataset, start_time=2, end_time=5):
    """Visualizes the waveforms for the training and testing datasets."""
    combined_waveform = []

    # Combine training dataset waveforms
    for entry in training_dataset:
        combined_waveform.extend(entry['stim_signals'])

    # Combine testing dataset waveforms
    for entry in testing_dataset:
        combined_waveform.extend(entry['question_stim'])

    # Determine the number of samples in one second
    samples_per_second = NEW_FS

    # Extract the samples corresponding to the specified time interval
    start_sample = start_time * samples_per_second
    end_sample = end_time * samples_per_second
    segment_waveform = combined_waveform[start_sample:end_sample]

    # Plot the specified segment of the waveform
    plt.figure(figsize=(22, 4), facecolor='black')
    plt.plot(segment_waveform, color='red', linewidth=0.5)
    plt.title(f'Waveform from {start_time} to {end_time} seconds', color='red', fontsize=10)
    plt.xlabel('Time (samples)', color='red', alpha=0.7)
    plt.ylabel('Microvolts', color='red', alpha=0.7)
    plt.xticks(color='red', alpha=0.7)
    plt.yticks(color='red', alpha=0.7)
    plt.gca().spines['bottom'].set_color('red')
    plt.gca().spines['left'].set_color('red')
    plt.gca().spines['bottom'].set_alpha(0.7)
    plt.gca().spines['left'].set_alpha(0.7)
    plt.gca().set_facecolor('black')
    plt.ylim(-1, 9)  # Adjust the limits as needed for your data

    plt.tight_layout()
    plt.show()

# Visualize waveforms
visualize_waveforms(alternating_dataset, testing_dataset, start_time=0, end_time=15)

# Print details from the training dataset
print("Training Dataset:")
for entry in alternating_dataset:
    print(f"Problem: {entry['problem']}, Binary Operand 1: {entry['binary_operand1']}, Binary Operand 2: {entry['binary_operand2']}, Binary Result: {entry['binary_result']}")
    print(f"Result: {entry['result']}, Correct: {entry['correct']}")
    print(f"Digital Signals: {entry['digital_signals']}")
    print(f"Outcome: {entry['outcome']}")
    print(f"Waveform: {entry['waveform']}")
    print(f"Stim Signals: {entry['stim_signals']}")
    print()

# Print details from the testing dataset
print("Testing Dataset:")
for entry in testing_dataset:
    print(f"Question: {entry['question']}, Binary Operand 1: {entry['binary_operand1']}, Binary Operand 2: {entry['binary_operand2']}")
    print(f"Digital Signals: {entry['digital_signals']}")
    print(f"Question Stim: {entry['question_stim']}")
    print()
```
